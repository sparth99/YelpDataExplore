```python
buisDF = pd.read_csv('yelp_business.csv')
```

```python
reviDF = pd.read_csv('yelp_review.csv')
```

```python
# merge the data set to get the latitude and longitude of each review with pandas merge function
df1 = reviDF.merge(buisDF[['business_id','latitude', 'longitude','city','state','name']], on=['business_id'])
```
